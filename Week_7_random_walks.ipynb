# Exercise 1

## 1)

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import random, seed

def walk(steps):
  x = np.zeros(steps, int)
  x[0] = 0
  for i in range(1, len(x)):
    rnd = random()
    if rnd < 0.5:
      x[i] = x[i-1] + 1
    else:
      x[i] = x[i-1] - 1
  return x

positions = walk(1000)
step_number = np.arange(1000)

plt.figure()
plt.plot(step_number, positions)
plt.xlabel('Step Number (i)')
plt.ylabel('Position (x)')
plt.title('Random Walk')
plt.grid()
plt.show()
```

## 2)

```python
plt.figure()

for _ in range(100):
  positions = walk(1000)
  step_number = np.arange(1000)
  plt.plot(step_number, positions)

plt.xlabel('Step Number (i)')
plt.ylabel('Position (x)')
plt.title('100 Random Walks')
plt.grid()
plt.show()
```

# Exercise 2

## 1)

```python
N = 10
steps = 1000

particles = np.zeros((steps, N), int)

for i in range(1, steps):
  for j in range(N):
    rnd = random()
    if rnd < 0.5:
      particles[i, j] = particles[i-1, j] + 1
    else:
      particles[i, j] = particles[i-1, j] - 1

plt.figure()
for j in range(N):
  plt.plot(step_number, particles[:, j])

plt.xlabel('Step Number (i)')
plt.ylabel('Position (x)')
plt.title('10 Random walks')
plt.grid()
plt.show()
```

## 2)

```python
N = 500
steps = 1000

particles = np.zeros((steps, N), int)

for i in range(1, steps):
  for j in range(N):
    rnd = random()
    if rnd < 0.5:
      particles[i, j] = particles[i-1, j] + 1
    else:
      particles[i, j] = particles[i-1, j] - 1

mean_position = np.mean(particles, axis=1)
mean_squared_position = np.mean(particles**2, axis=1)

last_step_index = steps - 1
mean_position_last_step = mean_position[last_step_index]
mean_squared_position_last_step = mean_squared_position[last_step_index]
print(f"Mean position at last step: {mean_position_last_step}")
print(f"Mean squared position at last step: {mean_squared_position_last_step}")

plt.figure()

for j in range(N):
  plt.plot(step_number, particles[:, j])

plt.xlabel('Step Number (i)')
plt.ylabel('Position (x)')
plt.title('500 Random walks')
plt.grid()
plt.show()
```

## 3)

```python
plt.figure()
plt.plot(step_number, mean_position)
plt.xlabel('Step Number (i)')
plt.ylabel('Mean Position (x)')
plt.title('Mean Position of 500 Random Walks')
plt.grid()
plt.axhline(y=0, color='r', linestyle='--', linewidth=1.0)
plt.show()
```

## 4)

```python
plt.figure()
plt.plot(step_number, mean_squared_position, label='Mean Squared Position')
plt.xlabel('Step Number (i)')
plt.ylabel('Mean Squared Position (x)')
plt.title('Mean Squared Position of 500 Random Walks')
plt.legend()
plt.grid()
plt.show()
```

I notice that the mean squared position increases linearly with th enumber of steps which makes sense logically

## 2)

```python
intermediate_steps = [10, 100, 500]
for step_index in intermediate_steps:
  positions_at_step = particles[step_index, :]
  plt.figure()
  plt.hist(positions_at_step, bins=50, density=True, alpha=0.7, color='b', label='Simulated Data')

  if step_index == 0:
    sigma = 1e-8
  else:
    sigma = np.sqrt(step_index)

  if np.isclose(x_min, x_max):
    x_min -= 3 * sigma
    x_max += 3 * sigma

  plt.xlabel('Position (x)')
  plt.ylabel('Probability Density')
  plt.title(f'Histogram of Particle Positions at Step {step_index}')
  plt.legend()
  plt.grid(axis='y', linestyle='--', alpha=0.7)
  plt.show()
```
